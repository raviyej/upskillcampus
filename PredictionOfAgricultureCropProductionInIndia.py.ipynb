```python
mr=mr.interpolate()
mr
```

```python
mr.isnull().sum()
```

```python
mr = mr.drop('Unnamed: 0', axis=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = mr[['Yield (Quintal/ Hectare) ','Cost of Cultivation (`/Hectare) A2+FL', 'Cost of Cultivation (`/Hectare) C2']]
Y = mr['Production 2006-07']

# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(X_train, Y_train)

# Make predictions on the testing set
Y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(Y_test, Y_pred)
rmse = mse**0.5
print("Root Mean Squared Error:", rmse)

# Use the model to make predictions on new data
a = input("Enter the State\n")
b = input("Enter the cultivation cost per hectare\n")
c = input("Enter the cultivation cost per hectare in c2\n")
new_data = pd.DataFrame({'Yield (Quintal/ Hectare) ': [a], 'Cost of Cultivation (`/Hectare) A2+FL': [b], 'Cost of Cultivation (`/Hectare) C2': [c]})
areaYield = model.predict(new_data)
print("the crop rate in 2010-11 is", areaYield)
```
